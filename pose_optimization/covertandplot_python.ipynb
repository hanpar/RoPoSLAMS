```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = './data/KittiGps_converted.txt'
i = 0
x = []
y = []
t = []
with open(filename) as f:
    for line in f:
        line_content = line.split(',')
        if i != 0:
            t.append(np.double(line_content[0]))
            x.append(np.double(line_content[1]))
            y.append(np.double(line_content[2]))
        i= i+1

plt.plot(x,y)
plt.show()
```

```python
inputgtfile = './data/floam_0027_gt_notsync.txt'
outputgtfile = './data/floam_gt_kitti_2011_09_30_0027_notsync.txt'
xgt = []
ygt = []
title = ["stamp","position.x","position.y","position.z","orientation.x","orientation.y","orientation.z","orientation.w"]
with open(outputgtfile, 'w') as fwrite:
    fwrite.write(' '.join(title))
    fwrite.write('\n')
    with open(inputgtfile,'r') as fread:
        next(fread)
        for line in fread:
            line_content = line.split(',')
            line_len = len(line_content)
            for i in range(4,line_len,10):
                fwrite.write(line_content[i+1])
                fwrite.write(' ')
                fwrite.write(' '.join(line_content[i+3:i+9]))
                fwrite.write('\n')
                xgt.append(np.double(line_content[i+3]))
                ygt.append(np.double(line_content[i+4]))
                    #[str(line_content[i]),str(line_content[i+1]),str(line_content[i+2]),str(line_content[i+3]),str(line_content[i+4]),
                    #str(line_content[i+5]),str(line_content[i+6]),str(line_content[i+7]),str(line_content[i+8]),str(line_content[i+9])])
plt.plot(xgt,ygt)
plt.show()  
```
